```
df_model = pd.DataFrame()
for i in list_offense_code_group:
  df_model = df_model.append(result_df.loc[result_df['OFFENSE_CODE_GROUP'] == i ])
  
```

```
factor_list = ["DISTRICT", "REPORTING_AREA", "YEAR", "MONTH","DAY_OF_WEEK", "HOUR", "UCR_PART","STREET","Lat","Long", "Day_Number"]
```

```
df_model.dtypes
```

```
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
```

```
le = LabelEncoder()
df_model["DISTRICT"] = le.fit_transform(df_model["DISTRICT"])
df_model["REPORTING_AREA"] = le.fit_transform(df_model["REPORTING_AREA"])
df_model["DAY_OF_WEEK"] = le.fit_transform(df_model["DAY_OF_WEEK"])
df_model["UCR_PART"] = le.fit_transform(df_model["UCR_PART"])
df_model["STREET"] = le.fit_transform(df_model["STREET"])
```

```
df_model
```

```
X = df_model[["DISTRICT", "REPORTING_AREA", "YEAR", "MONTH","DAY_OF_WEEK", "HOUR", "UCR_PART","STREET","Lat","Long", "Day_Number"]].values
Y = df_model[["OFFENSE_CODE_GROUP"]].values
```

```
Y = le.fit_transform(Y)
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
```

```
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.5, random_state = 121)
```

```
rf = RandomForestClassifier(n_estimators =2, max_depth =4,criterion ='entropy', random_state = 100)
rf.fit(X_train, Y_train)
pred = rf.predict(X_test) 
accuracy = accuracy_score(Y_test, pred)
```

```
accuracy
```

```
model = LogisticRegression()
model.fit(X_train, Y_train)
pred = model.predict(X_test) 
accuracy = accuracy_score(Y_test, pred)
```

```
print("Accuracy of the model: ", round(100 * f1_score(Y_test, pred, average='micro'),2), '%')
```

```
feature_importance = abs(model.coef_[0])
feature_importance = 100.0 * (feature_importance / feature_importance.max())
feature_importances = [(feature, round(importance, 2)) for feature, importance in zip(factor_list, feature_importance)]
feature_importances = sorted(feature_importances, key = lambda x: x[1], reverse = True)
indices = np.argsort(feature_importance)
imp = feature_importance
plt.figure(figsize=(10,5))
plt.title('Feature Importances')
plt.barh(range(len(indices)), imp[indices], color='b', align='center')
plt.yticks(range(len(indices)), [factor_list[i] for i in indices])
plt.xlabel('Relative Importance')
display()
```
